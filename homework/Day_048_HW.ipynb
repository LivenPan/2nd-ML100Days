```python
import os
import numpy as np
import pandas as pd
from datetime import date

from sklearn.model_selection import KFold, train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import datasets, metrics

DATA_ROOT = "../inputD48/"
```

```python
dfoff = pd.read_csv(os.path.join(DATA_ROOT,'train.csv'), header=None)
dftest = pd.read_csv(os.path.join(DATA_ROOT,'test.csv'), header=None)
dflabel = pd.read_csv(os.path.join(DATA_ROOT,'trainLabels.csv'), header=None)
# dftest = dftest[~dftest.Coupon_id.isna()]
# dftest.reset_index(drop=True, inplace=True)
print(dfoff.shape)
print(dftest.shape)
print(dflabel.shape)
dfoff.head(20)
```

```python
# 切分訓練集/測試集
x_train, x_test, y_train, y_test = train_test_split(dfoff, dflabel, test_size=0.25, random_state=42)

# 建立模型
reg = GradientBoostingClassifier(random_state=14)
```

```python
# 先看看使用預設參數得到的結果
reg.fit(x_train, y_train)
y_pred = reg.predict(x_test)
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
# 設定要訓練的超參數組合
n_estimators = [50, 100, 150]
max_depth = [1, 3, 5]
param_grid = dict(n_estimators=n_estimators, max_depth=max_depth)

## 建立搜尋物件，放入模型及參數組合字典 (n_jobs=-1 會使用全部 cpu 平行運算)
grid_search = GridSearchCV(reg, param_grid, scoring="neg_mean_squared_error", n_jobs=-1, verbose=1)

# 開始搜尋最佳參數
grid_result = grid_search.fit(x_train, y_train)

# 預設會跑 3-fold cross-validadtion，總共 9 種參數組合，總共要 train 27 次模型
```

```python
# 印出最佳結果與最佳參數
print("Best Accuracy: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
grid_result.best_params_
```

```python
# 使用最佳參數重新建立模型
reg_bestparam = GradientBoostingClassifier(max_depth=grid_result.best_params_['max_depth'],
                                           n_estimators=grid_result.best_params_['n_estimators'])

# 訓練模型
reg_bestparam.fit(x_train, y_train)

# 預測測試集
y_pred = reg_bestparam.predict(x_test)
```

```python
print(metrics.mean_squared_error(y_test, y_pred))
```

```python
# 預測結果
y_out = reg_bestparam.predict(dftest)
y_out = np.where(y_out >= 0.5, 1, 0)
print(y_out)
```

```python
fill = np.nan_to_num(dftest)
out = pd.DataFrame(y_out)

out.columns = ['Solution']
out['Id'] = np.arange(1, out.shape[0]+1)
out = out[['Id', 'Solution']]
out.head()
```
